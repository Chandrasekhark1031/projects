```python
plt.figure(figsize=(14, 8))
sns.countplot(y='AREA NAME', data=data, order=data['AREA NAME'].value_counts().index)
plt.title('Number of Crimes by Area')
plt.xlabel('Number of Crimes')
plt.ylabel('Area Name')
plt.show()
```

```python
plt.figure(figsize=(14, 8))
sns.countplot(x='Status Desc', data=data, order=data['Status Desc'].value_counts().index)
plt.xlabel('Status Desc')
plt.ylabel('Number of Crimes')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.histplot(data['Vict Age'].dropna(), bins=30, kde=True)
plt.title('Distribution of Victim Ages')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```
