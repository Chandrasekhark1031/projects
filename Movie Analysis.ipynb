```python
plt.figure(figsize=(10, 6))
sns.barplot(x='Genre', y='IMDb Rating', data=data)
plt.title('Rating by Genre ')
plt.xlabel('Genre')
plt.ylabel(' Rating')
plt.xticks(rotation=45)  # Rotate x labels if needed for better readability
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(x='Year', y='IMDb Rating', data=d)
plt.title('Rating by Year ')
plt.xlabel('Year')
plt.ylabel(' Rating')
plt.xticks(rotation=45)  
plt.show()
```

```python
# Calculate the counts for each visits
Genre_counts = data['Genre'].value_counts()

# Plot the pie chart
plt.figure(figsize=(10, 4))
plt.pie(Genre_counts, labels=Genre_counts.index, autopct='%1.1f%%', colors=sns.color_palette('muted'), startangle=140)
plt.title('Distribution of  Genre')
plt.show()
```

```python
plt.figure(figsize=(8, 4))
sns.histplot(data['Year'], bins=30, kde=True)
plt.title('Distribution of year')
plt.show()
```
