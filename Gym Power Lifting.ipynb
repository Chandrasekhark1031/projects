```python
# data visualization

sns.countplot(data=data, x="Lift Type")
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(x='Age', y='Amount Lifted (kg)', data=data, marker='o', color='b')
plt.title('Amount Lifted by Age')
plt.xlabel('Age')
plt.ylabel('Amount Lifted (kg)')
plt.grid(True)
plt.show()
```

```python
# Calculate the counts for each weight class
weight_class_counts = data['Weight Class'].value_counts()
# Plot the pie chart
plt.figure(figsize=(4, 4))
plt.pie(weight_class_counts, labels=weight_class_counts.index, autopct='%1.1f%%', colors=sns.color_palette('muted'), startangle=140)
plt.title('Distribution of Weight Classes')
plt.show()
```

```python
data.groupby('Weight Class').size().plot(kind='barh', color=sns.palettes.mpl_palette('Dark2'))
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
sns.scatterplot(x='Age', y='Amount Lifted (kg)', hue='Weight Class', data=data)
```

```python
sns.histplot(data['Amount Lifted (kg)'], bins=10, kde=True)
plt.title('Distribution of Amount Lifted')
plt.xlabel('Amount Lifted')
plt.ylabel('Frequency')
plt.show()
```

```python
# Calculate the counts for each Lifter name
Name_counts = data['Lifter Name'].value_counts()

# Plot the pie chart
plt.figure(figsize=(10,6))
plt.pie(Name_counts, labels=Name_counts.index, autopct='%1.1f%%', colors=sns.color_palette('muted'), startangle=140)
plt.title('Distribution of Lifter Names')
plt.show()
```
