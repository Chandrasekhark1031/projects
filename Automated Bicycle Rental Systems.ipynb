```python
sns.countplot(data=data, x="workingday")
```

```python
sns.lineplot(data=data,x='Vict Age',y='Weapon Desc')
```

```python
data.groupby('weathersit').size().plot(kind='barh', color=sns.palettes.mpl_palette('Dark2'))
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
year_counts = data['yr'].value_counts()
# Plot the pie chart
plt.figure(figsize=(4, 4))
plt.pie(year_counts, labels=year_counts.index, autopct='%1.1f%%', colors=sns.color_palette('muted'), startangle=140)
plt.title('Distribution of  year')
plt.show()
```

```python
plt.figure(figsize=(8, 4))
sns.histplot(data['temp'], bins=30, kde=True)
plt.title('Distribution of temp')
plt.show()
```
