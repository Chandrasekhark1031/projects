```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.barplot(x='Gender',y='LoyaltyProgram',data=data)
```

```python
data.groupby('ProductCategory').size().plot(kind='barh', color=sns.palettes.mpl_palette('Dark2'))
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
sns.lineplot(x="Age", y="AnnualIncome", data=data)
```

```python
plt.figure(figsize=(8, 4))
sns.boxplot(x='ProductCategory', y='NumberOfPurchases', data=data)
plt.title('NumberOfPurchases by Product Category')
plt.show()
```

```python
plt.figure(figsize=(8, 4))
sns.histplot(data['NumberOfPurchases'], bins=30, kde=True)
plt.title('Distribution of NumberOfPurchases')
plt.show()
```

```python
LoyaltyProgram_counts = data['LoyaltyProgram'].value_counts()
# Plot the pie chart
plt.figure(figsize=(4, 4))
plt.pie(LoyaltyProgram_counts, labels=LoyaltyProgram_counts.index, autopct='%1.1f%%', colors=sns.color_palette('muted'), startangle=140)
plt.title('Distribution of LoyaltyProgram')
plt.show()
```
