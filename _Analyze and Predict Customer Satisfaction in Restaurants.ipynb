```python
# Data Visualization
```

```python
data.groupby('FoodRating').size().plot(kind='barh', color=sns.palettes.mpl_palette('Dark2'))
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
satisfaction_counts = data['HighSatisfaction'].value_counts()

# Plot the pie chart
plt.figure(figsize=(4, 4))
plt.pie(satisfaction_counts, labels=satisfaction_counts.index, autopct='%1.1f%%', colors=sns.color_palette('muted'), startangle=140)
plt.title('Distribution of  satisfaction')
plt.show()
```

```python
sns.countplot(data=data, x='PreferredCuisine')
```

```python
# Calculate the counts for each visits
VisitFrequency_counts = data['VisitFrequency'].value_counts()

# Plot the pie chart
plt.figure(figsize=(4, 4))
plt.pie(VisitFrequency_counts, labels=VisitFrequency_counts.index, autopct='%1.1f%%', colors=sns.color_palette('muted'), startangle=140)
plt.title('Distribution of  VisitFrequency')
plt.show()
```

```python
plt.figure(figsize=(8, 4))
sns.lineplot(x='PreferredCuisine', y='AverageSpend', data=data, marker='o', color='b')
plt.title('AverageSpend by PreferredCuisine')
plt.xlabel('PreferredCuisine')
plt.ylabel('AverageSpend')
plt.grid(True)
plt.show()
```
